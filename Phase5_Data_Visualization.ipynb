```python
import joblib
```

```python
model = joblib.load("../models/spam_model.pkl")

print("Model Loaded Successfully!")
```

```python
vectorizer = joblib.load("../models/vectorizer.pkl")

print("Vectorizer Loaded Successfully!")
```

```python
message = input("Enter Email Message:\n\n")
```

```python
message_vector = vectorizer.transform([message])
```

```python
prediction = model.predict(message_vector)
```

```python
print("\n" + "="*50)
print("        EMAIL SPAM DETECTION")
print("="*50)

if prediction[0] == 1:
    print("🚨 EMAIL STATUS : SPAM")
    print("Warning! This email is Spam.")
else:
    print("✅ EMAIL STATUS : NOT SPAM")
    print("This email is Safe.")

print("="*50)
```

```python
while True:

    message = input("\nEnter Email Message (Type 'exit' to Quit):\n\n")

    if message.lower() == "exit":
        print("\nThank You!")
        break

    message_vector = vectorizer.transform([message])

    prediction = model.predict(message_vector)

    print("\n" + "="*50)

    if prediction[0] == 1:
        print("🚨 SPAM EMAIL")
    else:
        print("✅ NOT SPAM EMAIL")

    print("="*50)
```
