```python
import pandas as pd
```

```python
df = pd.read_csv("../dataset/emails.csv", encoding="latin-1")
```

```python
df = df[["v1", "v2"]]

df.columns = ["label", "message"]

df.head()
```

```python
df.columns = df.columns.str.strip()

print(df.columns.tolist())
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)

df.isnull().sum()
```

```python
print("Before :",df.shape)

df.drop_duplicates(inplace=True)

print("After :",df.shape)
```

```python
df["label"] = df["label"].map({
    "ham":0,
    "spam":1
})

df.head()
```

```python
df["message"] = df["message"].str.lower()

df.head()
```

```python
X = df["message"]

y = df["label"]
```

```python
print("Features :",X.shape)

print("Target :",y.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

X = cv.fit_transform(X)
```

```python
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("Training :",X_train.shape)

print("Testing :",X_test.shape)
```

```python
import joblib
import os

os.makedirs("../models",exist_ok=True)

joblib.dump(cv,"../models/vectorizer.pkl")

print("Vectorizer Saved Successfully")
```

```python
joblib.dump(X_train,"../outputs/X_train.pkl")
joblib.dump(X_test,"../outputs/X_test.pkl")
joblib.dump(y_train,"../outputs/y_train.pkl")
joblib.dump(y_test,"../outputs/y_test.pkl")

print("Processed Data Saved Successfully")
```
